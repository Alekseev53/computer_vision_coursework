```python
import numpy as np
import matplotlib.pyplot as plt

# Количество столбцов (бинов)
n_bins = 30

# Функция распределения вероятностей, например, нормальное распределение
# Среднее значение = 0, стандартное отклонение = 1, размер выборки = 1000
mean = 0
std_dev = 1
sample_size = 10

# Установка seed для генератора случайных чисел
seed = 42
np.random.seed(seed)

# Генерация данных
data = np.random.normal(mean, std_dev, sample_size)

# Создание гистограммы без нормализации для получения counts и bins
counts, bins, patches = plt.hist(data, bins=n_bins)

# Нормировка значений counts
counts_normalized = counts / sum(counts)

# Подготовка данных для отрисовки
bin_centers = 0.5 * (bins[:-1] + bins[1:])
width = np.diff(bins)

# Очистка предыдущей гистограммы
plt.clf()

# Отрисовка столбчатой диаграммы с нормированными значениями
plt.bar(bin_centers, counts_normalized, width=width, align='center')

# Настройка заголовка и осей
plt.title('Нормированная гистограмма распределения')
plt.xlabel('Значение')
plt.ylabel('Вероятность')

# Показать график
plt.show()

# Вывод значений бинов после нормировки
print("Значения бинов после нормировки:", counts_normalized)
print("Сумма значений бинов после нормировки:", sum(counts_normalized))
```
